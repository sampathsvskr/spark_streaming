```python
from kafka import KafkaProducer
import json
import uuid
import random
from datetime import datetime
```

```python
## Connect to kafka

BOOTSTRAP_SERVERS = 'kafka:29092'

def create_kafka_producer():

    # return kafka connection
    return KafkaProducer(
        bootstrap_servers = BOOTSTRAP_SERVERS,
        value_serializer=lambda v: json.dumps(v).encode('utf-8')
    )

def post_to_kafka(data):
    producer = create_kafka_producer()

    # send message to kafka
    producer.send(
        topic = 'test',
        key=bytes(str(uuid.uuid4()), 'utf-8'), 
        value=data
    )
    producer.close()
    
```

```python

# generate dummy data
def generate_event():
    devices = []
    num_devices = random.randint(1, 5)

    for i in range(num_devices):
        devices.append({
            "deviceId": f"D{str(i+1).zfill(3)}",
            "temperature": random.randint(-10, 50),
            "measure": "C",
            "status": random.choice(["SUCCESS", "ERROR"])
        })

    event = {
        "eventId": str(uuid.uuid4()),
        "eventOffset": random.randint(10000, 99999),
        "eventPublisher": random.choice(["device", "system"]),
        "customerId": f"CI{str(random.randint(1, 999)).zfill(5)}",
        "data": {
            "devices": devices
        },
        "eventTime": datetime.now().strftime("%Y-%m-%d %H:%M:%S.%f")
    }

    return event



```

```python

# write data to kafka
num_events = 2
for _ in range(num_events):
    data = generate_event()
    post_to_kafka(data)
    print('written to kafka')
    
```
